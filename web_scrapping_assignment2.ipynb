Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note: All of the above steps have to be done in code. No step is to be done manually.

```python
import selenium                                  #library that is used to work with selenium
from selenium import webdriver                   #importing webdriver module from selenium to open automated chrome window
import pandas as pd                              #to create DataFrame
from selenium.webdriver.common.by import By      #importing inbuilt class By 
import warnings                                  #to ignore any sort of warning
warnings.filterwarnings("ignore")
import time                                      #use to stop search engine for few seconds

driver = webdriver.Chrome()
driver.get("https://www.naukri.com/") #search for naukri.com in automated Chrome browser.
```

```python
#Entering the data like designation and location in the search fields
designation=driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys("Data Analyst")
```

```python
#Entering the location data in the search fields
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys("Bangalore")
```

```python
#Clicking on the Search button
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
#Creating empty lists for title, location, company name and exp
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#Scraping the job title from the given page and using for loop to append all the data into the empty list.
title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
```

```python
#Scraping the location from the given page and using for loop to append all the data into the empty list.
location_tags=driver.find_elements(By.XPATH,"//span[@class='ellipsis fleft locWdth']")
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)

#Scraping the company names from the given page and using for loop to append all the data into the empty list.    
company_tags=driver.find_elements(By.XPATH,"//a[@class='subTitle ellipsis fleft']")
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)

#Scraping the experience required from the given page and using for loop to append all the data into the empty list.
experience_tags=driver.find_elements(By.XPATH,"//span[@class='ellipsis fleft expwdth']")
for i in experience_tags[0:10]:
    exp=i.text
    experience_required.append(exp)
```

```python
#Checking if the length of all the lists is same to create a data frame.
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
driver.quit() #Close the automated chrome browser
```

```python
df= pd.DataFrame({'Title':job_title,'Location':job_location,'Company':company_name,'Experience':experience_required})
df
```

```python
import selenium                                  #library that is used to work with selenium
from selenium import webdriver                   #importing webdriver module from selenium to open automated chrome window
import pandas as pd                              #to create DataFrame
from selenium.webdriver.common.by import By      #importing inbuilt class By 
import warnings                                  #to ignore any sort of warning
warnings.filterwarnings("ignore")
import time                                      #use to stop search engine for few seconds

driver = webdriver.Chrome()
driver.get("https://www.flipkart.com/") #search for flipkart.com in automated Chrome browser.
```

```python
#To maximize the automated chrome window
driver.maximize_window()
```

```python
#Closing the sign in pop-up
close_popup = driver.find_element(By.XPATH,"/html/body/div[2]/div/div/button")
close_popup.click()
```

```python
#Entering the product in the search bar.
product=driver.find_element(By.CLASS_NAME,"_3704LK")
product.send_keys("sunglasses")
```

```python
#Click on the search button
search=driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search.click()
```

```python
#Creating empty list for each column in the data frame
brands = []
Product_Description = []
Price = []
```

```python
#Using if -else condition to reduce the repetition of the code.

if len(brands) < 100: # to check the length of the 'brands' list
    while True:
        brand_tags = driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')
        for i in brand_tags:
            brand=i.text
            brands.append(brand)
            if len(brands) >= 100:
                break  # Stop the loop if the desired number of brands is reached

        # Check if the button exists and click it if it does
        try:
            button = driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]') # to click on the next button
            button.click()
        except NoSuchElementException:
            break  # If the button does not exist, exit the loop
else:
    print(len(brands)) #Print the length of the list
```

```python
while len(Product_Description)<100: # to check the length of the 'Product_Description' list
    description_tags=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in description_tags:
        dscrpt=i.text
        Product_Description.append(dscrpt)
        if len(Product_Description)>=100:
            break # Stop the loop if the desired number of Product_Descriptions is reached
    # Check if the button exists and click it if it does        
    try:
        button = driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
        button.click()
    except NoSuchElementException:
        break # If the button does not exist, exit the loop
```

```python
print(len(Product_Description))
```

```python
while len(Price)<100: # to check the length of the 'Price' list
    price_tags=driver.find_elements(By.XPATH,'//div[@class="_25b18c"]')
    for i in price_tags:
        prices=i.text
        Price.append(prices)
        if len(Price)>=100:
            break #Stop the loop if the desired number of Price is reached
            
    # Check if the button exists and click it if it does  
    try:
        button = driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
        button.click()
    except NoSuchElementException:
        break # If the button does not exist, exit the loop
        
    
```

Q9: Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead,
Term of office, Remarks) from https://www.jagranjosh.com/.
This task will be done in following steps:
1. First get the webpagehttps://www.jagranjosh.com/
2. Then You have to click on the GK option
3. Then click on the List of all Prime Ministers of India
4. Then scrap the mentioned data and make theDataFrame.

```python
import selenium                                  #library that is used to work with selenium
from selenium import webdriver                   #importing webdriver module from selenium to open automated chrome window
import pandas as pd                              #to create DataFrame
from selenium.webdriver.common.by import By      #importing inbuilt class By 
import warnings                                  #to ignore any sort of warning
warnings.filterwarnings("ignore")
import time                                     #use to stop search engine for few seconds
import numpy as np                              #to create a DataFrame

driver = webdriver.Chrome()
driver.get("https://www.jagranjosh.com/") #Opening 'jagranjosh.com in automated chrome browser'
```

```python
#Clicking on the hamburger on the top left corner of the page.
hamburger = driver.find_element(By.CLASS_NAME,"mob-icon")
hamburger.click()
```

```python
#Slecting 'GK' from the list
GK = driver.find_element(By.XPATH,'//a[@href="/general-knowledge?ref=nav_dd"]')
GK.click()
```

```python
# Selecting the tab 'IndiaGK'
select_tab = driver.find_element(By.XPATH,'//a[@href="#indiaGK"]')
select_tab.click()
```

```python
#Selecting 'List of Prime_ministers'
name_tag = driver.find_element(By.XPATH,'//a[@href="/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1"]')
name_tag.click()
```

```python
pm_data = [] #creating empty list
pm_table = driver.find_element(By.CLASS_NAME, "table-box")
rows = pm_table.find_elements(By.TAG_NAME, "tr")[1:] #scrapping the data from row =1
```

```python
for row in rows:
    cols = row.find_elements(By.TAG_NAME, "td")
    name = cols[1].text.strip() #scapping the data in column one into 'name'
    born_dead = cols[2].text.strip() #scrapping the data in column two into 'born_dead'
    term_of_office = cols[3].text.strip() #scrapping the data in column three into 'Term_of_office'
    remarks = cols[4].text.strip()#scrapping data in column four into 'remarks'
    pm_data.append((name, born_dead, term_of_office, remarks)) #adding the data to the empty list created
```

```python
len(pm_data) #Length of the list 
```

```python
driver.quit() #Close the automated chrome browser
```

```python
data =np.array(pm_data)  #Using the NumPy array as the data is stored in array format
columns = ["Name", "Born-Dead", "Term of Office","Remarks"]
df = pd.DataFrame(pm_data, columns=columns)
df.index = range(1, len(df) + 1) #To start the indexing from 1 in dataframe
df
```

Q10: Write a python program to display list of 50 Most expensive cars in the world (i.e.
Car name and Price) from https://www.motor1.com/
This task will be done in following steps:
1. First get the webpage https://www.motor1.com/
2. Then You have to type in the search bar ’50 most expensive cars’
3. Then click on 50 most expensive cars in the world..
4. Then scrap the mentioned data and make the dataframe.

```python
import selenium                                  #library that is used to work with selenium
from selenium import webdriver                   #importing webdriver module from selenium to open automated chrome window
import pandas as pd                              #to create DataFrame
from selenium.webdriver.common.by import By      #importing inbuilt class By 
import warnings                                  #to ignore any sort of warning
warnings.filterwarnings("ignore")
import time                                      #use to stop search engine for few seconds

driver = webdriver.Chrome()
driver.get("https://www.motor1.com/") #Opening motor1 in automated chrome browser
```

```python
#To maximize the automated browser
driver.maximize_window()
```

```python
product = driver.find_element(By.XPATH,'/html/body/div[10]/div[2]/div/div/div[3]/div/div/div/form/input')
product.send_keys("50 most expensive cars") #Entering the query in the search bar.
```

```python
search = driver.find_element(By.XPATH,'//button[@class="m1-search-panel-button m1-search-form-button-animate icon-search-svg"]')
search.click() #Clicking onthe search button
```

```python
article = driver.find_element(By.XPATH,'/html/body/div[10]/div[9]/div/div[1]/div/div/div[2]/div/div[1]')
article.click() #Clicking on the article
```

```python
#Creating empty lists
cars = [] 
price = []
```

```python
#Scrapping the car names data from the article
cars_tags=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in cars_tags[0:50]:
    car=i.text
    cars.append(car) 
```

```python
#Scrapping  the price details from the article
price_tags = driver.find_elements(By.XPATH, '/html/body/div[10]/div[7]/div[2]/div[1]/div[2]/div[1]/p/strong')

for i in price_tags[0:50]:
    car_price = i.text
    price.append(car_price)
```

```python
#Checking the lenght of both the lists
print(len(cars),len(price))
```

```python
driver.quit() #Close the automated chrome browser
```

```python
#Creating dataFrame
df=pd.DataFrame({'Cars': cars,'price':price})
df
```
