Removing the outliers

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
threshold=3
print(np.where(z>3))
```

The above data shows the rows and columns that contains the high z score values, like 13th row 9th column and 14th row 5th column and so on

```python
z.iloc[13][9]
```

Will remove the data that contains outliers and create new data set


```python
df_new=df[(z<3).all(axis=1)]
```

```python
df_new
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

Importing the required libraried to divide the data into 2 parts, 70% for the training, i.e., the machine learns from the 70% of the data and then train from the rest 30% so that the classification report can be created

```python
x=df_new.iloc[:,0:-1] #the dataset is divided into the x and y, x contains all the attributes that effect the quality which is y
```

```python
y=df_new.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=47) 
#Splitting the data for the test and train, the test size should be 30 and random_state can be chaged and the result can be seen in the accuracy of the data
```

```python
x_train.shape 
```

```python
y_train.shape
```

The y train data contains 1015 rows and 1 column

```python
y_test.shape
```

The y test data contains 436 rows and 1 column

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train) #Naive Bayes method is used to predict the data
```

```python
pred=mnb.predict(x_test)
pred
```

```python
print(accuracy_score(y_test,pred)) #this is to check the accuracy of the report created. This result is 45% accurate
```

```python
print(confusion_matrix(y_test,pred))
```
