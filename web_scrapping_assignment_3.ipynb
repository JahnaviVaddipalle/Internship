Q4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
import selenium                                  #library that is used to work with selenium
from selenium import webdriver                   #importing webdriver module from selenium to open automated chrome window
import pandas as pd                              #to create DataFrame
from selenium.webdriver.common.by import By      #importing inbuilt class By 
import warnings                                  #to ignore any sort of warning
warnings.filterwarnings("ignore")
import time                                      #use to stop search engine for few seconds

driver = webdriver.Chrome()
driver.get('https://www.flipkart.com/')
```

```python
search = driver.find_element(By.XPATH,'//input[@class="Pke_EE"]')
search.send_keys('oneplus nord')
search.submit()
```

```python
brand,name,color = [],[],[]
for i in driver.find_elements(By.XPATH,'//div[@class = "_4rR01T"]'):
    brand.append(i.text.split(' ')[0])
    name.append(i.text)
    co = i.text.split('(')[-1]
    color.append(co.split(',')[0])
```

```python
ram,rom,pc,display,battery = [],[],[],[],[]
for i in driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]'):
    display.append(i.text.split('\n')[1])
    battery.append(i.text.split('\n')[-2])
    
    ra = i.text.split('\n')[0]
    ram.append(ra.split('|')[0])
    rom.append(ra.split('|')[1])
    
    cam = i.text.split('\n')[2]
    pc.append(cam.split('|')[0])
```

```python
price,link = [],[]
for i in driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]'):
    price.append(i.text)
for i in driver.find_elements(By.XPATH,"//a[@class = '_1fQZEK']"):
    link.append(i.get_attribute('href'))
```

```python
len(brand),len(name),len(color),len(ram),len(rom),len(pc),len(display),len(battery),len(price),len(link)
```

```python
df = pd.DataFrame({'Brand Name':brand,'Smartphone Name':name,'Color':color,'RAM':ram,'Storage(ROM)':rom,'Camera':pc,'Display':display})
```

Q7. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
hamburger = driver.find_element(By.CLASS_NAME,"_69hVhdY4")
hamburger.click()
time.sleep(2)
```

```python
billionaires = driver.find_element(By. XPATH, '//li[@class="TjJgrPSg cD45ib6e primary"]')
billionaires.click()
time.sleep(2)
```

```python
billionaires_data = []
```

```python
wait = WebDriverWait(driver, 10)
billionaire_table = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div[2]/div[1]/div[1]/div[1]/div[2]')))
rows = billionaire_table.find_elements(By.XPATH, '//div[@class="TableRow_cell__db-hv Table_cell__houv9"]')
```

```python
# Iterate through each row and extract the data
for i in range(0, len(rows), 7):
    rank = rows[i].text.strip()
    name = rows[i + 1].text.strip()
    net_worth = rows[i + 2].text.strip()
    age = rows[i + 3].text.strip()
    citizenship = rows[i + 4].text.strip()
    source = rows[i + 5].text.strip()
    industry = rows[i + 6].text.strip()
    
    billionaires_data.append((rank, name, net_worth, age, citizenship, source, industry))
```

```python
billionaires_data
```

```python
data = {
    'Rank': [entry[0] for entry in billionaires_data],
    'Name': [entry[1] for entry in billionaires_data],
    'Net Worth': [entry[2] for entry in billionaires_data],
    'Age': [entry[3] for entry in billionaires_data],
    'Citizenship': [entry[4] for entry in billionaires_data],
    'Source': [entry[5] for entry in billionaires_data],
    'Industry': [entry[6] for entry in billionaires_data]
}
```

```python
data =np.array(billionaires_data) 
columns = ["Rank", "Name", "Net Worth","Age","Citizenship","Source","Industry"]
```

Q9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
destination =driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input")
destination.send_keys("London")
```

```python
private_from_price = []
price = driver.find_elements(By.CLASS_NAME,"accommodation-price")
for i in price:
    private_from_price.append(i.text if i.text != "" else "-")
```

```python
private_from_price
```

```python
dorms = []
dorm = driver.find_elements(By.CLASS_NAME,"current")
for i in dorm:
    dorms.append(i.text if i.text != "" else "-")
```

```python
dorms
```

```python
property_description = []
description = driver.find_elements(By.XPATH, '//span[@class="distance-property-type"]')
for i in description:
    property_description.append(i.text if i.text != "" else "-")
```

```python
property_description
```

```python
wait = WebDriverWait(driver, 50)
button = driver.find_element(By.XPATH,'//button[@class="pill-content page-nav nav-right icon-only"]')
button.click() 
```

```python
while True:
    try:
        hotel = driver.find_elements(By.XPATH, '//div[@class="property-name"]')
        for i in hotel:
            hotels.append(i.text if i.text != "" else "-")

        dis = driver.find_elements(By.CLASS_NAME, "distance-description")
        for i in dis:
            distance.append(i.text if i.text != "" else "-")

        rating = driver.find_elements(By.CLASS_NAME, "number")
        for i in rating:
            ratings.append(i.text if i.text != "" else "-")

        review = driver.find_elements(By.CLASS_NAME, "review")
        for i in review:
            reviews.append(i.text if i.text != "" else "-")

        price = driver.find_elements(By.CLASS_NAME, "accommodation-price")
        for i in price:
            private_from_price.append(i.text if i.text != "" else "-")

        dorm = driver.find_elements(By.CLASS_NAME, "current")
        for i in dorm:
            dorms.append(i.text if i.text != "" else "-")

        description = driver.find_elements(By.XPATH, '//span[@class="distance-property-type"]')
        for i in description:
            property_description.append(i.text if i.text != "" else "-")

        wait = WebDriverWait(driver, 10)
        button = wait.until(EC.element_to_be_clickable((By.XPATH, '//button[@class="pill-content page-nav nav-right icon-only"]')))
        button.click()
    except NoSuchElementException:
        break
    except ElementClickInterceptedException:
        break
```

```python
len(hotel), len(distance), len(ratings), len(reviews), len(price), len(dorm), len(property_description)
```

```python
# Create a DataFrame from the collected data
data = {
    'Hostel Name': hotels[0:30],
    'Distance from City Centre': distance[0:30],
    'Ratings': ratings[0:30],
    'Total Reviews': reviews[0:30],
    'Privates from Price': private_from_price[0:30],
    'Dorms from Price': dorms[0:30],
    'Property Description': property_description[0:30]
}

df = pd.DataFrame(data)
df
```
